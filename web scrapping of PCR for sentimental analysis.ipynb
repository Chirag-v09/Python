```python
reviews = []
url = "https://niftyinvest.com/put-call-ratio/TATAMOTORS"
source_code = requests.get(url)
plain_text = source_code.text
soup = BeautifulSoup(plain_text, features="html.parser")
for re in soup.find_all('div', {'class': "row white darken-2 pcr-data-col z-depth-2"}):
    reviews.append(re.text)

oi_values = []
intraday_oi_values = []
for i in range(len(reviews)):
    reviews[i] = reviews[i].split('\n')
    if i == 0:
        for j in range(len(reviews[i])):
            if reviews[i][j] != '':
                oi_values.append(reviews[i][j])\
    
    if i == 1:
        for j in range(len(reviews[i])):
            if reviews[i][j] != '':
                intraday_oi_values.append(reviews[i][j])

pcr_oi = oi_values[-1]
pcr_intraday_oi = intraday_oi_values[-1]
print(pcr_oi, pcr_intraday_oi, sep = '\n')
```
